```python
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('Wednesday-workingHours.pcap_ISCX.csv')
```

```python
df
```

```python
#label encoding step
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()
df['Label ']= labelEncoder.fit_transform(df['Label '])
```

```python
df
```

```python
#Normalisation Start
#df1=df1.fillna(0)
df.replace([np.inf,-np.inf],np.nan,inplace=True)
df.fillna(1e6,inplace=True)
```

```python
#Normalisation(z-score) same concept followed by standard scaler!

from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#scaler.fit(df)



columns_to_normalize=df.columns[1:]
first_column=df[df.columns[0]]
scaler=StandardScaler()
df1_normalised= df.copy()
df1_normalised[columns_to_normalize]=scaler.fit_transform(df[columns_to_normalize])
df1_normalised[df.columns[0]]=first_column
```

```python
#scaled_data=scaler.transform(df)

df1_normalised
```

```python
#PCA ting

from sklearn.decomposition import PCA
pca = PCA(n_components=40)
pca.fit(df1_normalised)
```

```python
x_pca = pca.transform(df1_normalised)
```
