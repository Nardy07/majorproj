```python
df = df.drop(df[df['Label'].isin(['U2R','BFA','DoS'])].index)
```

```python
df=df.drop(columns=['Flow ID', 'Src IP','Dst IP','Timestamp'])
```

```python
#label encoding step
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()
X=df.drop("Label",axis=1)
y=df['Label'].values
from keras.utils import to_categorical
y_categorical = to_categorical(labelEncoder.fit_transform(y))
#y= labelEncoder.fit_transform(y)
```

```python
y_categorical
```

```python
#Normalisation(z-score) same concept followed by standard scaler!

from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#scaler.fit(df)



columns_to_normalize=X.columns[3:]
non_column=df[X.columns[0:2]]
scaler=StandardScaler()
X_scaled = X.copy()
X_scaled[columns_to_normalize]=scaler.fit_transform(X[columns_to_normalize])
X_scaled[df.columns[0:2]]=non_column
```

```python
X_scaled
```

```python
#PCA ting
from sklearn.decomposition import PCA
pca = PCA(n_components=48)
pca.fit(X_scaled)
```

```python
X_pca = pca.transform(X_scaled)
```

```python
X_reshaped = X_pca.reshape((X_pca.shape[0],1, X_pca.shape[1]))
```

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_categorical, test_size=0.2,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(units=64, input_shape=(X_reshaped.shape[1], X_reshaped.shape[2])))
model.add(Dense(3, activation='ReLU'))
```

```python
from sklearn.metrics import precision_score, recall_score
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history= model.fit(X_train, y_train, epochs=100, batch_size=128, validation_data=(X_test, y_test))
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
```

```python
import matplotlib.pyplot as plt
plt.plot(accuracy, label='Training Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
